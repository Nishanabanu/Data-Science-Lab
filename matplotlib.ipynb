```python
import matplotlib.pyplot as plt
import numpy as np

# Increase figure size
plt.figure(figsize=(12, 12))

# Graph 1
x = np.array([2, 4])
y = np.array([3, 6])

plt.subplot(3, 3, 1)
plt.title('Title 1')
plt.xlabel("X label", color="blue")
plt.ylabel("Y label", color="blue")
plt.plot(x, y, 'o', color="blue")

# Graph 2
y1 = np.array([3, 8, 1, 9])
y2 = np.array([6, 2, 8, 3])

plt.subplot(3, 3, 2)
plt.plot(y1, color="orange", marker='*')
plt.plot(y2, linewidth=1.5, color="blue", linestyle="dashed")

# Graph 3
ypoints = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='o', markersize=10, markeredgecolor="black", markerfacecolor="yellow")

# Graph 4
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([210, 220, 230, 240, 250, 260, 270, 280, 290, 300])

plt.subplot(3, 3, 4)
plt.title("Title 2")
plt.plot(x, y)
plt.grid()

# Graph 5
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
colors = np.random.randint(50, size=(50))
sizes =5 * np.random.randint(100, size=(50))

plt.subplot(3, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap='nipy_spectral')

# Graph 6
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 6)
plt.bar(x, y, color="red", width=0.5)  # Adjusted width for better visualization

# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
# Plot pie chart
plt.pie(y, labels=mylabels, explode=myexplode, startangle=90, shadow=True, autopct='%1.1f%%')
# Place legend outside the pie chart
plt.legend(title="Four Fruits:", bbox_to_anchor=(0.70, 1), loc='upper left')

# Graph 8
x = np.random.normal(100, 20, 200)
plt.subplot(3, 3, 8)
plt.hist(x, color='hotpink')

# Graph 9
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = ["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"]

plt.subplot(3, 3, 9)
plt.scatter(x, y, c=colors[:len(x)])

# Overall title
plt.suptitle("GRAPH", fontsize=20)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=1.3, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

# Show plot
plt.show()
```
